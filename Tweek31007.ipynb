```python
import numpy as np

# Create a 2x2 NumPy array
array_2x2 = np.array([[1, 2], [3, 4]])

print("1. Create a 2x2 NumPy array:")
print(array_2x2)
print("-" * 20)
```

```python
# Create a 3x NumPy array
# np.random.rand(rows, cols)
random_array_3x4 = np.random.rand(3, 4)

print("2. Create a 3x4 NumPy array:")
print(random_array_3x4)
print("-" * 20)
```

```python
# Create a NumPy array of all 1s(example 3x3)
ones_array = np.ones((3, 3))

# Create a NumPy array of all zeros (example 2x5)
zeros_array = np.zeros((2, 5))

print("3. Create a NumPy array of all 1s:")
print(ones_array)
print("\n3. Create a NumPy array of all zeros:")
print(zeros_array)
print("-" * 20)
```

```python
# Create a random 6x3 NumPy array
random_array_6x3 = np.random.randint(0, 10, size=(6, 3)) # 创建 0 到 9 的整数

# Transposed Array
transposed_array = random_array_6x3.T

print("4. Raw 6x3 random NumPy array (shape:", random_array_6x3.shape, "):")
print(random_array_6x3)
print("\n4. Transposed Array (shape:", transposed_array.shape, "):")
print(transposed_array)
print("-" * 20)
```

```python
# Set the dimension
n, p, m = 4, 3, 2

# Create the first array (nxp)
A = np.random.rand(n, p)

# Create the second array (pxm)
B = np.random.rand(p, m)

# Dot product (matrix multiplication)
# use np.dot(A, B) 或 A @ B
dot_product = A @ B

print(f"5. array A (n={n}, p={p}) The shape of: {A.shape}")
print(f"5. array B (p={p}, m={m}) The shape of: {B.shape}")
print("5. array A Dot product (matrix multiplication)  B:")
print(dot_product)
print("5. The shape of the dot product result:", dot_product.shape)
print("-" * 20)
```
